```python
#6. Array stacking and splitting
#Stacking means adding multiple array
#Two types of stacking are:
#1.vertical and 2.Horizontal
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.vstack(a))      #vstack is used for vertical stack-row wise
print(np.hstack(b))      #hstack is used for horizontal stack-column wise
```

```python
c=np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
split=np.hsplit(c,3)     #horizontal split

for s in split:
    print(s)
# print(np.hsplit(c,3))
```

```python
vsplit=np.vsplit(c,2)      #vertical split

for s in vsplit:
    print(s)
```

```python
#7. Mathematical operation on Array:
```

```python
a=np.array([10,20,50,-40])
print(a)
```

```python
print(a+10)   #add
print(a-30)   #subtract
print(a*2)    #multiplication
print(a/10)   #Divison
```

```python
b=np.array([1,4,7])
print(b)
```

```python
print(np.square(b))     #They square the number
print(np.sqrt(b))       # They give us square root of given number
```

```python
print(np.sin(b))
print(np.cos(b))
print(np.tan(b))
# print(np.sec(b))   #There is no such module in numpy
```

```python
#8.Mathematical operation on Multiple array:|
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
print(np.dot(a,b))    #dot product of given arr a and b
```

```python
print(a.T)   #Transpose
```

```python
t=np.array([[1,2,3],
           [4,5,6]])
print(t)
print(t.T)
```
